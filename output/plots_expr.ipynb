# Prostate plots

```python
fig, ax = plt.subplots(6,3)
fig.set_size_inches(20, 30)

#src_path = './eff_va_pros_00/'#'./ditto_grad_lrdc_1120/' #'./mri_ditto_1115/'
#src_path_list = ['./apros_cln_autoinn_01/', './apros_smis_autoinn_01/', './apros_mmis_autoinn_02/', './apros_hmis_autoinn_02/']
src_path_list = ['./apros_cln_autoinn_0417/', './apros_smis_autoinn_0417/', './apros_mmis_autoinn_0417/', './apros_hmis_autoinn_0417/']

client_name = ['BIDMC', 'BMC', 'HK', 'I2CVB', 'RUNMC', 'UCL']
label_list = ['clean', '40% mislabeled', '60% mislabeled', '80% mislabeled']

for s, src_path in enumerate(src_path_list):
    for c in range(6): #5
        with open(src_path+f'eval_result_client{c}.txt') as f:
            client = f.readlines()

        dice_list = []
        iou_list = []
        loss_list = []
        for i, strings in enumerate(client):
            dice = float(strings.split()[-3].split(':')[-1].split(',')[0])
            iou = float(strings.split()[-2].split(':')[-1].split(',')[0])
            loss = float(strings.split()[-1].split(':')[-1].split(',')[0])
            dice_list.append(dice)
            iou_list.append(iou)
            loss_list.append(loss)

        ax[c,0].plot(np.arange(1, len(dice_list)+1), dice_list, label=label_list[s])
        ax[c,0].set_xlabel('epochs')
        ax[c,0].set_ylabel('dice score')
        ax[c,0].set_yticks(np.arange(0,1.1,0.1))
        #ax[c,0].set_title(f'Client{c+1} Dice')
        ax[c,0].set_title(f'{client_name[c]} Dice')
        #ax[c,0].legend(bbox_to_anchor=(0.5, 0.5))

        ax[c,1].plot(np.arange(1, len(iou_list)+1), iou_list, label=label_list[s])
        ax[c,1].set_xlabel('epochs')
        ax[c,1].set_ylabel('IOU score')
        ax[c,1].set_yticks(np.arange(0,1.1,0.1))
        #ax[c,1].set_title(f'Client{c+1} IOU')    
        ax[c,1].set_title(f'{client_name[c]} IOU')
        #ax[c,1].legend(bbox_to_anchor=(0.5, 0.5))
        
        ax[c,2].plot(np.arange(1, len(loss_list)+1), loss_list, label=label_list[s])
        ax[c,2].set_xlabel('epochs')
        ax[c,2].set_ylabel('Val loss')
        ax[c,2].set_yticks(np.arange(0,2.1,0.2)) #0-2, 0.2
        #ax[c,2].set_title(f'Client{c+1} val loss')
        ax[c,2].set_title(f'{client_name[c]} Val loss')
        ax[c,2].legend(bbox_to_anchor=(1, 0.5))
        
#fig.tight_layout()
store_path = './paper_plots/'
#fig.suptitle('Prostate dataset validation result plots (\u03B3-logistic dice loss)', fontsize=20)
plt.savefig(f'{store_path}pros_gamma_0420.png', bbox_inches='tight')
#plt.savefig('./mri_ditto_1115/ditto_1115')
plt.show()
```
